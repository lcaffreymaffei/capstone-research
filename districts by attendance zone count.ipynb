```python
wide = pd.read_csv('Full Segregation Index.csv', header=0)
wide['geoleaid'] = wide['geoleaid'].astype(str)
wide.loc[wide['geoleaid'].str.len() == 6, 'geoleaid'] = '0' + wide['geoleaid']
```

```python
wide = wide[['geoleaid', 'n_attendancezones00', 'n_attendancezones10', 'n_attendancezones11', 'n_attendancezones12',
            'n_attendancezones14', 'n_attendancezones16']]
```

```python
wide.loc[(wide.n_1z_yrs != wide.n_yrs) & (wide.n_1z_yrs != 0)][0:50]
```

```python
# add number of years districts have 1 attendance zone
wide['n_1z_yrs'] = 0
wide.loc[wide.n_attendancezones00 == 1, 'n_1z_yrs'] = wide['n_1z_yrs'] + 1
wide.loc[wide.n_attendancezones10 == 1, 'n_1z_yrs'] = wide['n_1z_yrs'] + 1
wide.loc[wide.n_attendancezones11 == 1, 'n_1z_yrs'] = wide['n_1z_yrs'] + 1
wide.loc[wide.n_attendancezones12 == 1, 'n_1z_yrs'] = wide['n_1z_yrs'] + 1
wide.loc[wide.n_attendancezones14 == 1, 'n_1z_yrs'] = wide['n_1z_yrs'] + 1
wide.loc[wide.n_attendancezones16 == 1, 'n_1z_yrs'] = wide['n_1z_yrs'] + 1
```

```python
# add number of years districts in dataset
wide['n_yrs'] = 0
wide.loc[pd.notna(wide.n_attendancezones00), 'n_yrs'] = wide['n_yrs'] + 1
wide.loc[pd.notna(wide.n_attendancezones10), 'n_yrs'] = wide['n_yrs'] + 1
wide.loc[pd.notna(wide.n_attendancezones11), 'n_yrs'] = wide['n_yrs'] + 1
wide.loc[pd.notna(wide.n_attendancezones12), 'n_yrs'] = wide['n_yrs'] + 1
wide.loc[pd.notna(wide.n_attendancezones14), 'n_yrs'] = wide['n_yrs'] + 1
wide.loc[pd.notna(wide.n_attendancezones16), 'n_yrs'] = wide['n_yrs'] + 1
```

```python
# classify districts as either having 1 attendance zone or > 1 attendance zone
wide['az_count'] = '1'
wide.loc[(wide.n_1z_yrs == 0) | (wide.n_1z_yrs / wide.n_yrs <= 0.5), 'az_count'] = '> 1'
```
